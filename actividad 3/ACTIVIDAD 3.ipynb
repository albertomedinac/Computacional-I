```python
#importamos a panda
import pandas as pd
import numpy as np
```

```python
#descargamos los datos
df=pd.read_csv('aguaprieta2',skiprows=20,skipfooter=1, sep='\s+',header=None,
               names=['DD/MM/YY','PRECIP','EVAP','TMAX','TMIN'],engine='python')
```

```python
df.dtypes
```

```python
df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['DD/MM/YY'], 1), dayfirst=True)
df=df.drop(['DD/MM/YY'],1)
```

```python
df['MES'] = df['FECHA'].dt.month
```

```python
df['AÑO']=df['FECHA'].dt.year
```

```python
#aqui hacemos un for loop donde segun el mes nos da la suma
for i in range(1,13):
        PRECIPMENS =  df.loc[df['MES'] == i, 'PRECIP'].sum()
        print('MES',i,np.round(PRECIPMENS,decimals=1),'mm')
```

```python
#esto es para sumar los datos de una columna donde cumple con la condicion de otra columna
df.loc[df['MES'] == 2, 'PRECIP'].sum()
```

```python
#aqui hacemos un for loop donde segun el mes nos da la suma
for i in range(1,13):
        TMINMENS =  df.loc[df['MES'] == i, 'TMIN'].min()
        print('MES',i,np.round(TMINMENS,decimals=1),'C')
```

```python
#aqui hacemos un for loop donde segun el mes nos da la suma
for i in range(1,13):
        TMAXMENS =  df.loc[df['MES'] == i, 'TMAX'].max()
        print('MES',i,np.round(TMAXMENS,decimals=1),'C')
```

```python
#aqui hacemos un for loop donde segun el mes nos da la suma
for i in range(1961,2012):
        EVAPAÑO =  df.loc[df['AÑO'] == i, 'EVAP'].sum()
        print('AÑO',i,np.round(EVAPAÑO,decimals=1),'')
        
```
