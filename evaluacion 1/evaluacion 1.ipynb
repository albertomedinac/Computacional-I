```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('kino.dat')
```

```python

df.dtypes
```

```python
df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['TIMESTAMP'], 1), dayfirst=True)
df = df.drop(['TIMESTAMP'], 1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['MES']=df['FECHA'].dt.month
df['SEG']=df['FECHA'].dt.second
df['DIA']=df['FECHA'].dt.day
df['MIN']=df['FECHA'].dt.minute
df['HORA']=df['FECHA'].dt.hour
df['AÑO']=df['FECHA'].dt.year
```

```python

df['TEMP']=df['AirTC_Avg']
df.drop(['AirTC_Avg'],1)
df.head()
```

```python
#hacer
df3=df[["FECHA","SEG","MIN","HORA","DIA","MES","AÑO","TEMP"]]
```

```python
df3=df3.loc[df['FECHA']>='2018-11-01 00:00:00']
```

```python
df3['TPROM']=df3.groupby(['AÑO','MES','DIA','HORA'])['TEMP'].transform('mean')
```

```python
df3 = df3[df3["MIN"]==0]
```

```python
df3.reset_index(inplace=True)
#df3=df3.drop(df3['level_0','index'])
df3.head()
```

```python
UF24=[]
count1=0
for i in range(0,len(df3.TPROM)):
    if(df3["TPROM"][i]<=1.5):
        count1=0     
    elif(df3["TPROM"][i]>1.5 and df3["TPROM"][i]<=2.5):
        count1=0.5
    elif(df3["TPROM"][i]>2.5 and  df3["TPROM"][i]<=9.2):
        count1=1.0
    elif(df3["TPROM"][i]>9.2 and df3["TPROM"][i]<=12.5):
        count1=0.5
    elif(df3["TPROM"][i]>12.5 and df3["TPROM"][i]<=16):
        count1=0
    elif(df3["TPROM"][i]>16.0 and df3["TPROM"][i]<=18.0):
        count1=-0.5
    elif(df3["TPROM"][i]>18):
        count1=-1
        
    UF24.append(count1)
    count1=0

df3['UFH'] =UF24    
df3.head()
```

```python
df3['UFD'] = df3.groupby(['AÑO','MES','DIA'])['UFH'].transform('sum')
df3.head()
```

```python
df3.head()
```

```python
#agrupamos los maximos y minimos
df3['TMAXD']=df3.groupby(['AÑO','MES','DIA'])['TPROM'].transform('max')
df3['TMIND']=df3.groupby(['AÑO','MES','DIA'])['TPROM'].transform('min')

```

```python
df3.head()
```

```python
df4=df3[df3['HORA']==0]
```

```python
#df4=df4.drop(['index','SEG','MIN'],axis = 1)
```

```python
df4.reset_index(inplace=True)
df4.head()
```

```python
df4=df4.drop(['index','MIN','HORA','SEG'],axis=1)
```

```python
df4.head()
```

```python
df4.tail()
```

```python
#ahora graficas por dia la temperatura máxima y mínima
df5 = df4[['TMAXD','TMIND']]
plt.figure(); df5.plot(); plt.legend(loc='best')
plt.title("Evolución de la temperatura de cada dia")
plt.ylabel("Temp ºC")
plt.xlabel("dias")
plt.grid(True)
plt.savefig("1",plt=2000)
plt.show()
```
