```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('kino.dat')
```

```python
df.dtypes
```

```python
df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['TIMESTAMP'], 1), dayfirst=True)
df = df.drop(['TIMESTAMP'], 1)
```

```python
df.head()
```

```python

df['MES']=df['FECHA'].dt.month
df['SEG']=df['FECHA'].dt.second
df['DIA']=df['FECHA'].dt.day
df['MIN']=df['FECHA'].dt.minute
df['HORA']=df['FECHA'].dt.hour
df['AÃ‘O']=df['FECHA'].dt.year
```

```python
df['TEMP']=df['AirTC_Avg']
df.drop(['AirTC_Avg'],1)
df.head()
```

```python
#hacer
df3=df[["FECHA","SEG","MIN","HORA","DIA","MES","AÃ‘O","TEMP"]]
```

```python
df3=df3.loc[df['FECHA']>='2018-11-01 00:00:00']
```

```python
df3['TPROM']=df3.groupby(['AÃ‘O','MES','DIA','HORA'])['TEMP'].transform('mean')
```

```python
df3 = df3[df3["MIN"]==0]
```

```python
df3.reset_index(inplace=True)
#df3=df3.drop(df3['level_0','index'])
df3.head()
```

```python
UF24=[]
count1=0
for i in range(0,len(df3.TPROM)):
    if(df3["TPROM"][i]<=1.5):
        count1=0     
    elif(df3["TPROM"][i]>1.5 and df3["TPROM"][i]<=2.5):
        count1=0.5
    elif(df3["TPROM"][i]>2.5 and  df3["TPROM"][i]<=9.2):
        count1=1.0
    elif(df3["TPROM"][i]>9.2 and df3["TPROM"][i]<=12.5):
        count1=0.5
    elif(df3["TPROM"][i]>12.5 and df3["TPROM"][i]<=16):
        count1=0
    elif(df3["TPROM"][i]>16.0 and df3["TPROM"][i]<=18.0):
        count1=-0.5
    elif(df3["TPROM"][i]>18):
        count1=-1
        
    UF24.append(count1)
    count1=0

df3['UFH'] =UF24    
df3.head()
```

```python
df3['UFD'] = df3.groupby(['AÃ‘O','MES','DIA'])['UFH'].transform('sum')
df3.head()
```

```python
HF = []
HC = []

for i in range(df3["AÃ‘O"][0],df3["AÃ‘O"][len(df3)-1]+1):
    dfnen = df3[df3["AÃ‘O"] == i]
    for j in range(dfnen["MES"][dfnen.index[0]],dfnen["MES"][dfnen.index[len(dfnen)-1]]+1):
        dfmes = dfnen[dfnen["MES"] == j]
        for k in range(dfmes["DIA"][dfmes.index[0]],dfmes["DIA"][dfmes.index[len(dfmes)-1]]+1):
            dfdia = dfmes[dfmes["DIA"] == k]
            for n in dfdia.index:
                if(0 < dfdia["TPROM"][n] <= 10):
                    HF.append(1)
                    HC.append(0)
                elif(dfdia["TPROM"][n] > 25):
                    HF.append(0)
                    HC.append(1)
                else:
                    HF.append(0)
                    HC.append(0)
```

```python
df3["HF"] = HF
df3["HC"] = HC
```

```python
df3.head()
```

```python
df3["HF24"] = df3.groupby(["AÃ‘O","MES","DIA"])["HF"].transform("sum")
df3["T>25"] = df3.groupby(["AÃ‘O","MES","DIA"])["HC"].transform("sum")
df3["HFE"] = df3["HF24"] - df3["T>25"]
```

```python
df3.head()
```

```python
df4=df3[df3["HORA"]==0]
```

```python
df4.reset_index(inplace=True)
df4.head()
```

```python
df4=df4.drop(['index'],axis=1)
```

```python
df4.head()
```

```python
# GrÃ¡fica de Temperatura y Humedad Relativa
df5 = df4[['HF24','HFE']]
plt.figure(); df5.plot(); plt.legend(loc='best')
plt.title("ComparaciÃ³n de la HF24 y HFE")
plt.ylabel("")
plt.grid(True)
plt.show()
```
